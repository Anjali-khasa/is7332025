```python
import yfinance as yf
stock_data_hourly = yf.download(tickers=tickers, period='2y', interval='1h')  
print(f"\n shape: {stock_data_hourly.shape}\n\n cols: \n{stock_data_hourly.columns}\n ")
print(f"\n {stock_data_hourly.head(10)}")
print(f"\n {stock_data_hourly.tail(10)}")
```

```python
import yfinance as yf
# Intraday data (intervals less than "1d") is only available for the last 60 days 
stock_data_15min = yf.download(tickers=tickers, period="60d", interval="15m") 
print(f"\n shape: {stock_data_15min.shape}\n\n cols: \n{stock_data_15min.columns}\n ")
print(f"\n {stock_data_15min.head(10)}")
print(f"\n {stock_data_15min.tail(10)}")
```

```python
# Read data from Yahoo Finance
def read_data(tickers, start_date, end_date):
    df = yf.download(tickers=tickers, start=start_date, end=end_date, progress=True)
    print(f"\n shape: {df.shape}\n\n cols: \n{df.columns}\n {df.head(10)}")
    return df
df_prices = read_data(tickers, start_date, end_date)
```

```python
# Flatten the multi-index: move 'Ticker' from columns to rows
df_flat = df_prices.stack(level=1).reset_index()
df_flat.columns
df_flat.head(10)

# Restructure and compute target columns
# df = df_prices['Open'].stack().reset_index()

# df.columns = ['Date', 'Ticker', 'Open']
# df['Close'] = df_prices['Close'].stack().values

# # Add other features
# for col in ['High', 'Low', 'Volume']:
#     df[col] = df_prices[col].stack().values
```

# create target column

```python
# Step 2: Create the target column with next day's open price
df_flat['target_next_close'] = df_flat.groupby('Ticker')['Open'].shift(-1)
df_flat.head(10)

# df = df.sort_values(['Ticker', 'Date']).reset_index(drop=True)
# df['target_next_close'] = df.groupby('Ticker')['Close'].shift(-1)
```

```python
df_flat.tail(5)
```

# Data cleaning 

```python
# Filter out rows where next_open is NaN (i.e., last record for each ticker)
# df_cleaned = df_flat[df_flat['target_next_open'].notna()].copy()
# Drop rows with missing target
df_cleaned = df_flat.dropna(subset=['target_next_close'])

# Reset index if needed
df_cleaned.reset_index(drop=True, inplace=True)
df_cleaned.head(10)
# # Show result
# print(df_cleaned[['Date', 'Ticker', 'Open', 'next_open']])
```

```python
df_cleaned['Ticker'].value_counts()
```

```python
df_cleaned = df_cleaned[["Date", "Ticker", "Volume", "High","Low", "Open", "Close", "target_next_close"]]
```

# add more features

```python
# Daily Return
df_cleaned['daily_return'] = (df_cleaned['Close'] - df_cleaned['Open']) / df_cleaned['Open']

# Moving Averages
df_cleaned['moving_avg_5'] = df_cleaned['Close'].rolling(window=5).mean()
df_cleaned['moving_avg_10'] = df_cleaned['Close'].rolling(window=10).mean()

# Volatility (Rolling Standard Deviation of Close)
df_cleaned['volatility_5'] = df_cleaned['Close'].rolling(window=5).std()

# Lag Features of `target_next_close` (simulate this column first)
df_cleaned['target_next_close'] = df_cleaned['Close'].shift(-1)
df_cleaned['target_next_close_lag1'] = df_cleaned['target_next_close'].shift(1)
df_cleaned['target_next_close_lag2'] = df_cleaned['target_next_close'].shift(2)
```

```python
df_cleaned = df_cleaned.dropna()
# df_cleaned = df_flat.dropna(subset=['target_next_close'])
df_cleaned.head(100)
```
