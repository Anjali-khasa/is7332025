```python
# Read data from Yahoo Finance
def read_data(tickers, start_date, end_date):
    df = yf.download(tickers=tickers, start=start_date, end=end_date, progress=True)
    print(f"\n shape: {df.shape}\n\n cols: \n{df.columns}\n {df.head(10)}")
    return df

tickers = ['AAPL', 'MSFT', 'AMZN', 'GOOG'] # , 'sp500'
# AAPL	BA	T	MGM		IBM	TSLA
start_date = '2010-01-01'
end_date = '2025-01-01'
df_prices = read_data(tickers, start_date, end_date)
```

#### When dealing with time series data, during the train test spilt there are various ways of doing that- cross validation or directly using ramdom seeds(works very well with linear regression or classification problem). With LSTM we divide the traning data ( should be ordered with respective of date) 70% and test 30% based on date only as next day data is dependent on the previous day. 

```python
# Split the data into training and testing sets
traning_size= int(len(df1) * 0.65)
test_size = len(df1) - traning_size
train_data, test_data = df1[0:traning_size,:], df1[traning_size:len(df1),:1]
```

#### data preprocessig - we have to take timesteps (independent features) x_train = input that we will give (dependent feature)  Y_train = out put we will get based on the input. This will used to train the data. We will do the same for test date also. So after train LSTM model with training data set, will take this test data set and predicit it and then try to compare with Y_test and then we will se how good accuracy will be comming and how good forecasting is actually come. 

#### F1 to F100 so here it will take f1 to f99 in X_train and f100 will be the output, then it will again take f2 to 100 and out put will be taken as f 101 and continously we will have the output 

```python
import numpy
# Convert an array of values into a dataset matrix
def create_dataset(dataset, time_step=1):
    dataX, dataY = [], []
    for i in range(len(dataset)-time_step-1):
        a = dataset[i:(i+time_step), 0] ### i = 0, 1, 2, 3
        dataX.append(a)
        dataY.append(dataset[i + time_step, 0])
    return numpy.array(dataX), numpy.array(dataY)
```

```python
# reshape into X=t , t+1 , t+2, t+3 and Y=t+4
time_step = 100
X_train, y_train = create_dataset(train_data, time_step)
X_test, y_test = create_dataset(test_data, time_step)
```

#### Before implementing any lstm we need to reshape our X_train into 3d, just add 1 which will convert into 3D because we will be giving this as an input to our lstm 

```python
# resape the input to be [samples, time steps, features] which is required for LSTM
X_train = X_train.reshape(X_train.shape[0], X_train.shape[1], 1)
X_test = X_test.reshape(X_test.shape[0], X_test.shape[1], 1)
```

```python
# Create the Stacked LSTM model
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import LSTM
```

```python
model = Sequential()
model.add(LSTM(50, return_sequences=True, input_shape=(100, 1)))
model.add(LSTM(50, return_sequences=True))
model.add(LSTM(50))
model.add(Dense(1))
model.compile(loss='mean_squared_error', optimizer='adam')
model.summary()
```

```python
model.fit(
    X_train,
    y_train,
    validation_data=(X_test, y_test),  
    epochs=100,
    batch_size=64,
    verbose=1
)
```
